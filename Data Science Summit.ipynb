```python
from datetime import date
import plotly.express as px
import plotly.graph_objects as go
from app.utils import as_table
today = date(2020, 11, 21)
```

```python
countries = [[i.country_or_region, i.province_or_state, i.active, i.deaths, i.confirmed, i.latitude, i.longitude] for i in DailyData.objects.filter(date=today, deaths__gt=0, active__gt=0, confirmed__gt=0)]
```

```python
countries_asmap = [{"lat": i[-2], "lng": i[-1], "deaths": i[-3]} for i in countries]
```

```python
px.density_mapbox(countries_asmap, lat="lat", lon="lng", z="deaths", radius=10, mapbox_style="stamen-terrain", center={"lat": 0, "lon": 180}, zoom=0)
```

```python
bangladesh = DailyData.objects.filter(country_or_region__icontains="canada").order_by("date").values("date", "deaths", "confirmed", "active")
```

```python
bangladesh
```

```python
x = [i["date"] for i in bangladesh]
```

```python
y1 = [i["confirmed"] for i in bangladesh]
y2 = [i["active"] for i in bangladesh]
y3 = [i["deaths"] for i in bangladesh]
```

```python
fig = px.bar(bangladesh, x = x, y = y1)
fig.add_trace(go.Bar(x = x, y = y2))
fig.add_trace(go.Bar(x = x, y = y3))
fig.show()
```
